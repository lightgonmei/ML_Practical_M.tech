# Q1.Importing Dataset of our choice and displaying the first five rows, and print the datatype of the DataFrame as a whole.And also displaying the statistical summary (mean, median, standard deviation, etc.) for each column and provide complete information about the DataFrame.


```python
import pandas as pd
from google.colab import drive
```

```python
# 2. Define Headers
# The raw dataset lacks headers, so we define them manually to prevent
# the first row of data from being misinterpreted as column names.
headers_data = ["symboling", "normalized-losses", "make", "fuel-type", "aspiration",
                "num-of-doors", "body-style", "drive-wheels", "engine-location",
                "wheel-base", "length", "width", "height", "curb-weight", "engine-type",
                "num-of-cylinders", "engine-size", "fuel-system", "bore", "stroke",
                "compression-ratio", "horsepower", "peak-rpm", "city-mpg", "highway-mpg", "price"]

print("Headers type:", type(headers_data))
```

```python
# 3. Read the Dataset
# We pass 'names=headers_data' to assign the columns we defined above.
file_path = '/content/drive/MyDrive/ML Lab practice/1.Importing Datasets/dataset_1.data'
df = pd.read_csv(file_path, names=headers_data)
```

```python
# 4. Data Inspection (First 5 Rows)
print("\n--- The first 5 rows of the dataframe ---")
print(df.head(5))
```

```python
# 5. Data Inspection (Last 10 Rows)
print("\n--- The last 10 rows of the dataframe ---")
print(df.tail(10))
```

```python
# 6. Check Data Types
print("\n--- Data Types of each column ---")
print(df.dtypes)
```

```python
# 7. Statistical Summary
print("\n--- Statistical Summary (Include All) ---")
# include="all" ensures we see stats for non-numeric columns (like 'make' or 'fuel-type')
print(df.describe(include="all"))
```
