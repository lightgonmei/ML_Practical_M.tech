# Q5. Applying EM algorithm to cluster a set of data stored in a .CSV file. Use the same data set for clustering using k-Means algorithm. And comparing the results of these two algorithms and giving a comment on the quality of clustering.

```
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score
```

```
# Step 1: Load the Iris dataset
# We use the built-in Iris dataset which mimics loading structured data from a CSV
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```
# Step 2: Cluster the data using k-Means algorithm
# k-Means requires us to specify the number of clusters (3 for Iris)
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans_labels = kmeans.fit_predict(X)
kmeans_silhouette = silhouette_score(X, kmeans_labels)
```

```
# Step 3: Cluster the data using EM algorithm (Gaussian Mixture Model)
# GMM also requires the number of components
gmm = GaussianMixture(n_components=3, random_state=42)
gmm_labels = gmm.fit_predict(X)
gmm_silhouette = silhouette_score(X, gmm_labels)
```

```
# Step 4: Compare the clustering results and visualize them
print(f"Silhouette Score for k-Means: {kmeans_silhouette:.4f}")
print(f"Silhouette Score for EM (GMM): {gmm_silhouette:.4f}")
# Visualization of the clusters
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))
# k-Means Clustering Plot
ax1.scatter(X[:, 0], X[:, 1], c=kmeans_labels, cmap='viridis', marker='o', edgecolor='k', s=50)
ax1.set_title('k-Means Clustering')
ax1.set_xlabel('Sepal Length')
ax1.set_ylabel('Sepal Width')
# EM (GMM) Clustering Plot
ax2.scatter(X[:, 0], X[:, 1], c=gmm_labels, cmap='viridis', marker='o', edgecolor='k', s=50)
ax2.set_title('EM (GMM) Clustering')
ax2.set_xlabel('Sepal Length')
ax2.set_ylabel('Sepal Width')
plt.tight_layout()
plt.show()
# Step 5: Comment on the quality of clustering
if kmeans_silhouette > gmm_silhouette:
    print("\nConclusion: k-Means clustering provides better quality clusters according to the silhouette score.")
else:
    print("\nConclusion: EM (GMM) clustering provides better quality clusters according to the silhouette score.")
```
