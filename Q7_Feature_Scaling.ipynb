# Q7. Python program to load (or create) a dataset, split it into training and testing sets, and apply feature scaling to the data. Finally, display the dataset before and after scaling to observe the changes.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
# 1. Loading the Dataset
df = pd.read_csv("/content/drive/MyDrive/ML Lab practice/7 Feature Scaling & Data Normalization/Dataset_03.csv")
print("--- Original Data Head ---")
print(df.head())
print("\n" + "="*50 + "\n")
```

```python
# 2. Splitting Features and Target
X = df.iloc[:, :-1]  # All columns except the last one
y = df.iloc[:, -1]   # The last column
```

```python
# 3. Splitting into Train and Test Sets
# random_state=1 ensures reproducible output
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
X_train = X_train.copy()
X_test = X_test.copy()
```

```python
# 4. Feature Scaling
sc = StandardScaler()
```

```python
# Apply fit_transform on Training data (calculates mean/std AND scales)
# We apply this ONLY to columns from index 4 onwards (Age and Salary)
print("--- Scaling Columns at index 4 and beyond ---")
X_train.iloc[:, 4:] = sc.fit_transform(X_train.iloc[:, 4:])
```

```python
# Apply transform on Test data (uses mean/std from Training data)
X_test.iloc[:, 4:] = sc.transform(X_test.iloc[:, 4:])
```

```python
# 5. Display Results
print("\n--- X_train (After Scaling) ---")
print(X_train)
print("\n--- X_test (After Scaling) ---")
print(X_test)
```
