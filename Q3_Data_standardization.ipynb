# Q3. Python program to load (or create) a dataset, split it into training and testing sets, and standardize the features using StandardScaler from sklearn. preprocessing. Finally, display the dataset before and after standardization to observe the changes.

```python
import pandas as pd
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# 1. Loading the Dataset
# We use the Breast Cancer dataset available in sklearn
from sklearn.datasets import load_breast_cancer
dataset=load_breast_cancer()
print (dataset)
```

```python
# Loading the data to a pandas dataframe
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
```

```python
# Display first 5 rows and shape
print("--- First 5 rows of the DataFrame ---")
print(df.head())
print("\nShape of the DataFrame:", df.shape)
```

```python
# 2. Separating Features (X) and Target (Y)
x = df
y = dataset.target
```

```python
print("\n--- Target Variable (Y) ---")
print(y)
```

```python
# 3. Splitting the data into Training and Test data
# Using 80% for training and 20% for testing, with random_state=3 for reproducibility
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=3)
```

```python
print("\n--- Shapes of Split Data ---")
print(f"Original X: {x.shape}")
print(f"X Train: {x_train.shape}")
print(f"X Test: {x_test.shape}")
```

```python
# 4. Standardization
# Check the standard deviation of the original data to see if scaling is needed
print(f"\nStandard Deviation of original data: {dataset.data.std()}")
```

```python
# Initialize StandardScaler
scaler = StandardScaler()
```

```python
# Fit the scaler on the Training data
scaler.fit(x_train)
```

```python
# Transform (Standardize) both Training and Test data
x_train_standardized = scaler.transform(x_train)
x_test_standardized = scaler.transform(x_test)
```

```python
# 5. Display Results
print("\n--- Standardized Training Data (First 3 rows) ---")
print(x_train_standardized[:3])
```

```python
print("\n--- Standard Deviation after Standardization ---")
# Ideally, this should be close to 1.0
print(x_train_standardized.std())
```

```python
print("\n--- Standardized Test Data (First 3 rows) ---")
print(x_test_standardized[:3])
```
