# Q4 Python program to load (or create) a dataset, separate it into predictor variables (X) and target variable (y), and then split the dataset into training and testing sets. Finally, print both the training and testing datasets.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
# 1. Importing the Dataset
df = pd.read_excel("/content/drive/MyDrive/ML Lab practice/3 & 4 Data standardization and Train & test data splitting/Salary_Dataset_02.xls")
print("--- First 5 rows of the dataset ---")
print(df.head())
print("\n" + "="*50 + "\n")
```

```python
# 2. Dividing into Target and Predictor Variables
# X stores predictor variables (All columns except the last one)
# y stores target variable (The last column)
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
print("--- Predictors (X) ---")
print(X.head())
print("\n--- Target (y) ---")
print(y.head())
print("\n" + "="*50 + "\n")
```

```python
# 3. Splitting into Training and Testing Sets
# test_size = 0.2 means 20% of data is for testing
# random_state = 90 ensures consistent splitting results
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=90)
```

```python

# 4. Displaying the Output
print("--- X_train (Training Predictors) ---")
print(X_train)
print("\n--- X_test (Testing Predictors) ---")
print(X_test)
print("\n--- y_train (Training Targets) ---")
print(y_train)
print("\n--- y_test (Testing Targets) ---")
print(y_test)
```
